```python
import pandas
import matplotlib.pyplot as plt
import gmplot #installation in readme
import numpy as np
import collections
from wordcloud import WordCloud #installation in readme
from PIL import Image
import networkx as nx
import os
```

```python
os.chdir('..')
terror=pandas.read_csv("Terror_data.csv",encoding = "ISO-8859-1")
os.chdir('./UvA-IV2017-Team9/')
```

```python
plt.scatter(terror.longitude[:],terror.latitude[:])
plt.show()
```

```python
terrorHeatmap=terror.dropna(subset=['longitude'])
terrorHeatmap=terror.dropna(subset=['latitude'])
```

```python
gmap1 =  gmplot.GoogleMapPlotter(0,0,3)

gmap1.heatmap(terrorHeatmap.latitude[:],terrorHeatmap.longitude[:],radius=30)
gmap1.draw("terror.html")
```

```python
counter = dict(collections.Counter(terrorHeatmap['country_txt']).most_common(20))

country_names = counter.keys()
country_counts = counter.values()

indexes = np.arange(len(country_names))
width = 0.7
plt.bar(indexes, country_counts, width)
plt.xticks(indexes + width * 0.5, country_names,rotation=90)
plt.show()
```

```python
counter = dict(collections.Counter(terrorHeatmap['city']).most_common(20))

city_names = counter.keys()
city_counts = counter.values()

indexes = np.arange(len(city_names))
width = 0.7
plt.bar(indexes, city_counts, width)
plt.xticks(indexes + width * 0.5, city_names,rotation=90)
plt.show()
```

```python
counter = dict(collections.Counter(terrorHeatmap['iyear']))

year_names = counter.keys()
year_counts = counter.values()

indexes = np.arange(len(year_names))
width = 0.7
plt.bar(indexes, year_counts, width)
plt.xticks(indexes + width * 0.5, year_names,rotation=90)
plt.show()
```

```python
counter = dict(collections.Counter(terrorHeatmap['gname']).most_common(20))

gname_names = counter.keys()
gname_counts = counter.values()

indexes = np.arange(len(gname_names))
width = 0.7
plt.bar(indexes, gname_counts, width)
plt.xticks(indexes + width * 0.5, gname_names,rotation=90)
plt.show()
```
